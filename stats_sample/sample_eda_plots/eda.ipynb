```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df
```

### Distribution plots

### histgram

```python
df['Type 1'].value_counts().plot.bar()
plt.show()
```

```python
df['Type 2'].value_counts().sort_values(ascending=True).plot.barh()
plt.show()
```

#### distplot

```python
sns.distplot(df['Speed'])
plt.show()
```

```python
sns.kdeplot(df['HP'])
plt.show()
```

### Relational plots

#### numeric

```python
plt.scatter(x=df['Attack'], y=df['Defense'])
plt.show()
```

```python
sns.relplot(x='Sp. Atk', y='Sp. Def', hue='Generation', data=df)
plt.show()
```

#### Continuous variables

```python
x = np.linspace(0, 10, num=100)
```

```python
y = np.sin(x)
sns.lineplot(x, y)
plt.show()
```

```python
y = np.exp(x)
sns.lineplot(x, y)
plt.show()
```

### Categorical plots

```python
sns.swarmplot(x='HP', y='Type 1', data=df)
plt.show()
```

```python
sns.boxplot(x='Attack', y='Type 2', data=df)
plt.show()
```

```python
sns.violinplot(x='Generation', y='Defense', data=df)
plt.show()
```

```python
sns.boxenplot(x='Legendary', y='Sp. Atk', data=df)
plt.show()
```

### Matrix plots

#### Categorical variables & Categorical variables

```python
table_df = df[['Type 1', 'Type 2']].pivot_table(index='Type 1', columns='Type 2', aggfunc=len)
```

```python
sns.heatmap(table_df, cmap='Blues', linewidth=.1)
plt.show()
```

```python
sns.clustermap(table_df.fillna(0), cmap='Blues', linewidth=.1)
plt.show()
```
