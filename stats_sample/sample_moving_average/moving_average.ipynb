### n-th moving average

fomula

$ N = 2k+1 $

$$ ma_i = \frac{1}{n} \{ x_i + \sum^k_{i=1} (x_{i-j} + x_{i+j}) \} $$

$ N = 2k $

$$ ma_i = \frac{1}{n} \{ x_i + 0.5 \times (x_{i-k} + x_{i+k}) + \sum^{k-1}_{i=1} (x_{i-j} + x_{i+j}) \} $$
